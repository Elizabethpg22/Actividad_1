```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Metodo por columnas
#filtro3 = a1.iloc[: , 1:3] #Filas consecutivas
filtro3 = a1.iloc[: , [4,7,10]] #Filas no consecutivas
filtro3
```

```python
a2 =pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col="NOMBRE_VENDEDOR")
a2
```

```python
#Aplicar filtro filas vs columnas
filtro4=a2.loc[["LETICIA RAMIREZ HERNANDEZ", "DIEGO ARMANDO LEON DURAN"], ["DESCR"]]
filtro4
```

```python
filtro4.to_csv("Filtro filas vs columnas.csv")
from google.colab import files
files.download("Filtro filas vs columnas.csv")
```

```python
#Filtro de cabecera
filtro5=a2.head(10)
filtro5
```

```python
filtro5.to_csv("Filtro de cabecera.csv")
from google.colab import files
files.download("Filtro de cabecera.csv")
```

```python
#Filtro por comparación
filtro6 = a2[a2["COSTO_UNITARIO"] > 7000]
filtro6
```

```python
filtro6.to_csv("Filtro por comparación.csv")
from google.colab import files
files.download("Filtro por comparación.csv")
```

```python
#Filtro Y
filtro7 = a2[(a2["PRECIO_UNITARIO"]<6000) & (a2["FECHA_DOC"]=="2022-02-28")]
filtro7
```

```python
filtro7.to_csv("Filtro  Y.csv")
from google.colab import files
files.download("Filtro  Y.csv")
```

```python
#Filtro OR
filtro8 = a2[(a2 ["PRECIO_UNITARIO"] < 10)| (a2 ["FECHA_DOC"]=="2022-12-23")]
filtro8
```

```python
filtro8.to_csv("Filtro  OR.csv")
from google.colab import files
files.download("Filtro  OR.csv")
```

```python
#filtro not
filtro9 = a2[~(a2['PRECIO_UNITARIO'] < 102) & ~(a2['FECHA_DOC'] == '2022-12-23')]
filtro9
```

```python
filtro9.to_csv("Filtro  not.csv")
from google.colab import files
files.download("Filtro  not.csv")
```
