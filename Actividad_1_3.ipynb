```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
c1 =pd.read_excel('Inmuebles.xlsx',skiprows= 1)
c1
```

a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”

```python
filtro1 = c1.drop(['Referencia'], axis = 1)
filtro1
```

```python
filtro1.to_csv("Filtro 1 Act3.csv")
from google.colab import files
files.download("Filtro 1 Act3.csv")
```

```python
c1.info()
```

b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00

```python
filtro2 = c1[(c1['Fecha_Alta'] == '2007-04-29 00:00:00') | (c1['Fecha_Alta'] == '2004-01-02 00:00:00')]
filtro2
```

```python
filtro2.to_csv("Filtro 2 Act3.csv")
from google.colab import files
files.download("Filtro 2 Act3.csv")
```

c) Todos los registros de “ Tipo: parking, local, industrial y piso”

```python
filtro3 = c1[(c1 ["Tipo"]== "Parking") | (c1 ["Tipo"]== "Local") | (c1 ["Tipo"]== "Industrial") | (c1 ["Tipo"]== "Piso")]
filtro3

```

```python
filtro3.to_csv("Filtro 3 Act3.csv")
from google.colab import files
files.download("Filtro 3 Act3.csv")
```

d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”

```python
filtro4 = c1[(c1["Provincia"] == 'Girona') & (c1["Superficie"] >100)]
filtro4
```

```python
filtro4.to_csv("Filtro 4 Act3.csv")
from google.colab import files
files.download("Filtro 4 Act3.csv")
```

e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”

```python
filtro5 = c1[(c1["Precio Venta"] > 1000000) | (c1["Vendedor"] == 'Carmen') | (c1["Vendedor"] == 'Joaquín')]
filtro5
```

```python
filtro5.to_csv("Filtro 5 Act3.csv")
from google.colab import files
files.download("Filtro 5 Act3.csv")
```

f) Todos los registros que no sean de “Provincia : Barcelona“

```python
filtro6 = c1[~(c1['Provincia'] == 'Barcelona')]
filtro6
```

```python
filtro6.to_csv("Filtro 6 Act3.csv")
from google.colab import files
files.download("Filtro 6 Act3.csv")
```

g)Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“

```python
filtro7 = c1[~(c1['Operación'] == 'Alquiler') & ~(c1['Tipo'] == 'Oficina')]
filtro7
```

```python
filtro7.to_csv("Filtro 7 Act3.csv")
from google.colab import files
files.download("Filtro 7 Act3.csv")
```
