```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
b1 =pd.read_excel('Datos de Facturaci√≥n.xlsx')
b1
```

```python
b1.info()
```

```python
b1.CVE_CLPV=b1.CVE_CLPV.replace({'O':0}, regex=False)
b1.CVE_CLPV=b1.CVE_CLPV.replace({'6781|':'6781'}, regex=False)
b1.CVE_CLPV=b1.CVE_CLPV.replace({'3612C':'3612'}, regex=False)
b1.CVE_CLPV=b1.CVE_CLPV.replace({'      ':'0'}, regex=False)
b1.CVE_CLPV=b1.CVE_CLPV.replace({'2385A':'2385'}, regex=False)
b1.CVE_CLPV=b1.CVE_CLPV.replace({'2510A':'2510'}, regex=False)
b1.CVE_CLPV=b1.CVE_CLPV.replace({'8440Q':'8440'}, regex=False)
b1.CVE_CLPV=b1.CVE_CLPV.replace({'M':'0'}, regex=False)
```

```python
b1['CVE_CLPV']=b1['CVE_CLPV'].astype(float)
```

```python
filtro1 = b1[(b1['CVE_CLPV'] >= 1000) & (b1['CVE_CLPV'] <= 2000)]
filtro1
```

```python
filtro1.to_csv("Filtro 1.csv")
from google.colab import files
files.download("Filtro 1.csv")
```
